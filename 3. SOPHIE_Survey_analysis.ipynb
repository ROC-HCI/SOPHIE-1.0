```python
def reduce_cols(df):
  column_names = list(df.columns)
  column_dict = {}
  for i in range(len(column_names)):
    new_name = "c" + str(i+1)
    column_dict[column_names[i]] = new_name
  df = df.rename(columns=column_dict)
  # invert key, value in column_dict
  column_dict = {v: k for k, v in column_dict.items()}
  return df, column_dict
```

```python
survey_control, column_dict_control = reduce_cols(survey_control)
survey_sophie, column_dict_sophie = reduce_cols(survey_sophie)
survey_uiux, column_dict_uiux = reduce_cols(survey_uiux)
```

# Merge with results

```python
# prompt: in survey_control dataframe, column 'c6', drop all rows where 'c6' is NaN. Turn the remaining values to lower case. If the values are not in the format p followed by one or two digits, drop that row as well.

import pandas as pd
import re

def clean_df(df, column_dict): # 1. Drop rows where 'c6' is NaN df = df.dropna(subset=['c6'])
  # 2. Convert remaining values to lower case
  df['c6'] = df['c6'].str.lower()

  # 3. Keep rows where 'c6' matches 'p' followed by one or two digits
  pattern = r'^p\d{1,2}$'
  df = df[df['c6'].str.match(pattern, na=False)]

  # 4. Remove the 'p' and keep the integer
  def extract_number(text):
      match = re.search(r'\d+', text)
      return int(match.group(0)) if match else None

  df['c6'] = df['c6'].apply(extract_number)
  return df


survey_control = clean_df(survey_control, column_dict_control)
survey_sophie = clean_df(survey_sophie, column_dict_sophie)
```

```python
survey_control = survey_control[['c6', 'c7', 'c8', 'c9', 'c10']]
survey_sophie = survey_sophie[['c6', 'c7', 'c8', 'c9', 'c10']]
```

```python
# Rename columns in survey_control
survey_control = survey_control.rename(columns={
    'c6': 'P_ID',
    'c7': 'GENUINE_EMPATHY_FOR_SP',
    'c8': 'HARD_TO_FEEL_REAL_EMPATHY',
    'c9': 'HARD_TO_EXPRESS_REAL_EMPATHY',
    'c10': 'APPEARENCE_CAN_BE_TAOUGHT'
})

# Rename columns in survey_sophie
survey_sophie = survey_sophie.rename(columns={
    'c6': 'P_ID',
    'c7': 'GENUINE_EMPATHY_FOR_SP',
    'c8': 'HARD_TO_FEEL_REAL_EMPATHY',
    'c9': 'HARD_TO_EXPRESS_REAL_EMPATHY',
    'c10': 'APPEARENCE_CAN_BE_TAOUGHT'
})
```
